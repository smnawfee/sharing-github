```python
import pandas as pd
import io #to load uploaded data
from bertopic import BERTopic
from sklearn.feature_extraction.text import CountVectorizer #to remove stopwords
```

Run Bertopic for RBKC

```python
#load data
from google.colab import files
uploaded = files.upload()
```

```python
#load wiki data for kc
kc_wiki_with_summary= pd.read_csv(io.BytesIO(uploaded['kc_wiki_with_summary (1).csv']))
summary_kc = kc_wiki_with_summary['page_summary'].tolist()
```

Process text and run model

```python
#remove stopwords
cv = CountVectorizer(stop_words="english")
```

```python
#run model
topic_model = BERTopic(language="english", calculate_probabilities=True, verbose=True, vectorizer_model= cv)
topics, probs = topic_model.fit_transform(summary_kc)
```

Visualise topics

```python
topic_model.get_topic_info()
```

```python
topic_model.get_topics()
```

```python
topic_model.get_document_info(summary_kc)
```

```python
topic_model.visualize_barchart(top_n_topics=5)
```

```python
topic_model.generate_topic_labels()
```
